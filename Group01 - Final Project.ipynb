Build a program that can be used as a basic calculator. Your program should have a menu displayed for the user to choose from, where are listed basic operations: addition, subtraction, multiplication, division, second power, square root, exit.
The program should allow the user to choose the desired operation over and over again until user chooses to quit using

```python
import math
# Step 1: print a simple menu
# Step 5: to repeat the menu until user exit we only need to add while in the beginning, and it makes the whole proccess unavailable. Here we need indentations. we should push all lines to function as a subcommand.
while True:
    print("===== BASIC \033[1mGroup ONE\033[0m CALCULATOR =====")
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")
    print("5. Second power (x^2)")
    print("6. Square root (âˆšx)")
    print("7. Exit")
# We need this to choose from operations / It only shows the list of options on the screen. There are no inputs yet.
# To make the title bold we used  \033[1mGroup ONE\033[0m : from BOLD ="\033[1m" to END ="\033[0m"

# Step 2: get a single menu choice
    choice = input("Choose an option (1-7): ")
    print("You chose:", choice)

# Step 3: doing the operation, let's test only addition
## a = float(input("Enter first number: "))
## b = float(input("Enter second number: "))
## print("Result:", a + b)
# but it is a one function command, we should add if/elif/else to add more operations

# Step 4: let's add other operations
# Step 4_1: We want it also shows the numbers
    if choice == "1":
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print('Addition')
        print(a)
        print(b)
        print("Result:", a + b)

    elif choice == "2":
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print('Subtraction')
        print(a)
        print(b)
        print("Result:", a - b)

    elif choice == "3":
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print('Multiplication')
        print(a)
        print(b)
        print("Result:", a * b)

    elif choice == "4":
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        if b != 0:
            print('Division')
            print(a)
            print(b)
            print("Result:", a / b)

        else:
            print("Error: Division by zero is not allowed.")
            print('division by zero is not allowed.')
            print(a)
            print(b)

    elif choice == "5":
        a = float(input("Enter a number: "))
        print('Second power')
        print(a)
        print("Result:", a ** 2)

    elif choice == "6":
        a = float(input("Enter a number: "))
        if a >= 0:
            print('square root')
            print(a)
            print("Result:", math.sqrt(a))
        else:
            print(a)
            print("Error: Cannot take square root of negative number.")

    elif choice == "7":
        print("Goodbye ðŸ‘‹")
        print("It was a \033[1mGroup ONE\033[0m production")
        break # to quit after adding 7

    else:
        print("Invalid choice.")

# Step 5: to repeat the menu until user exit we only need to add while in the beginning
# Step 7: to end the process by adding the 7 value, There is a command we should use named: break
```
